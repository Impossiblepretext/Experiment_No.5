```python
''' Problem statement 1: unique event participation
A college event has a list of students who registered multiple times by mistake
write a program using set to store participant names and display only unique participants'''
# List of students who registered (with duplicates)
participants = [
    "Aman", "Riya", "Sahil", "Riya", "Neha",
    "Aman", "Karan", "Neha"
]

# Using a set to store unique participant names
unique_participants = set(participants)

# Display unique participants
print("Unique Event Participants:")
for name in unique_participants:
    print(name)
```

```python
''' Problem statement 2:Three students choose  elective subjects
Find the common subjects chosen by both students using set operations.
Concept used: Set interaactions ()'''
# Lists of subjects
vibhas = ["maths", "physics", "chemistry"]
gitesh = ["maths", "physics", "biology"]
anay = ["business", "maths", "biology"]

# Convert lists to sets
s1 = set(vibhas)
s2 = set(gitesh)
s3 = set(anay)

# Common subjects chosen by all three students
common_subjects = s1 & s2 & s3

print("Students:")
print("Vibhas:", vibhas)
print("Gitesh:", gitesh)
print("Anay:", anay)

print("\nCommon subject(s) chosen by all students:")
print(common_subjects)

```

```python
''' Problem statement 3: one set contains studdent in cricket club, another in football club .
Find students who are:
1. In both clubs
2. In only one club
cooncept used : Inetrsection,symmetric difference '''
# Students in clubs
cricket = {"Aman", "Rohit", "Vibhas", "Anay"}
football = {"Gitesh", "Rohit", "Anay", "Karan"}

# 1. Students in both clubs (Intersection)
both_clubs = cricket & football

# 2. Students in only one club (Symmetric Difference)
only_one_club = cricket ^ football

print("Students in both Cricket and Football clubs:")
print(both_clubs)

print("\nStudents in only one club:")
print(only_one_club)
```

```python
''' Problem Statement 4: Remove Invalid Enteries.
A set contains valid course codes.
Remove invalid codes from the from the set.
Concept Used: remove()/discard()'''
courses={"ETC101","ETC102","ETC103"}
courses.discard("ETC101")
print(courses)
```

```python
# Basics Of Dictionary
laptop = {
    "brand" : "Dell",
    "model" : "G15",
    "year" : 2022
}

print(laptop)

# Length of the dictionary
print(len(laptop))

# Type of variable
print(type(laptop))
# Output :

# Changing the values
laptop["year"] = 2023
print(laptop)

# Retrival of values
print("Brand : ",laptop['brand'])

# Adding a new item with kay
laptop["Price"] = 55000
print(laptop)

# Removing elements from list
laptop.pop("Price") # Price & 55000 removed from dictionary
print(laptop)
```
